```python
%%time

processed_docs = list()
num = 0
for doc in nlp.pipe(texts[:10],n_process=6,batch_size = 100):
    ents = doc.ents # name entities
    
    # Lemmatize token, remove punctutation and stopwords
    doc = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop]
    
    doc.extend([str(entity) for entity in ents if len(entity) > 1])
    processed_docs.append(doc)
    print(num+1)

pickle.dump(processed_docs, open("datasets_output/texts/processed_docs_cord19_scispacy.pk", "wb"))
```
