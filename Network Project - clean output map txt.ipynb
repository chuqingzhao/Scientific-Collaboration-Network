```python
def node_word(node_dict,text):
    new_dict = {}
    for key,value in node_dict.items():
        if key<=len(text):
            row = text['id'][key-1]
            if row < 5401:
                word = text['label'][row-1]
            else:
                #print(key)
                word = text['label'][row-3]
            new_dict[word]=value
        else:
            row = text['id'][key-2]
            word = text['label'][row-2]
            new_dict[word]=value
            #print(row,word)
    return new_dict
```
