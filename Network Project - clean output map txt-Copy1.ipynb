# Network of March Data

```python
import networkx as nx
import sklearn #For generating some matrices
import pandas #For DataFrames
import numpy as np #For arrays
import matplotlib.pyplot as plt #For plotting
import seaborn #Makes the plots look nice
```

```python
g = nx.Graph()
```

```python
node_1 = text_march_network[0]
```

```python
node_list1 = [i for i in text_march_network[0]]
node_list2 = [i for i in text_march_network[1]]
```

```python
len(node_list1)
```

```python
node = (node_list1,node_list2)
```

```python
for i in range(len(node_list1)):
    g.add_node(node_list1[i])
    g.add_node(node_list2[i])
    g.add_edge(node_list1[i],node_list2[i])
```

```python
len(g.nodes)
```

```python
len(g.edges)
```

```python
print(nx.info(g))
```

```python
dcentralities = nx.degree_centrality(g)
sorted(dcentralities.items(), reverse = True, key = lambda x:x[1])
```
